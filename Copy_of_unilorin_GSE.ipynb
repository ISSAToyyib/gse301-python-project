Task 2.1 : Conditional statement for grading

```python
def grade_score(score):
    if score >= 70:
        return "A"
    elif score >= 60:
        return "B"
    elif score >= 50:
        return "C"
    elif score >= 45:
        return "D"
    elif score >= 40:
        return "E"
    else:
        return "F"

mark = int(input("Enter your score: "))
grade = grade_score(mark)

match grade:
    case "A":
        print("Excellent performance")
    case "B":
        print("Very good")
    case "C":
        print("Good")
    case _:
        print("Needs improvement")


```

Part 4 : INTERACTIVE MENU SYSTEM

```python
# Task 4.1: Build a console menu
students = ["Toyyib","Tunde","Zaynab"]

while True:
    print("WELCOME TO STUDENT MANAGEMENT MENU")
    print("1. View all students")
    print("2. Add new student")
    print("3. Check eligibility for graduation")
    print("4. Find top performance")
    print("5. Exit")

    option = int(input("Select an option (1-5): "))

    match option:
        case 1:
            if not students:
                print("No students available.")
            else:
                print("\nList of Students:")
                for student in students:
                    print(student)

        case 2:
            name = input("Enter student name: ")
            score = float(input("Enter student score: "))
            students.append({"name": name, "score": score})
            print("Student added successfully.")

        case 3:
            if not students:
                print("No students available.")
            else:
                for student in students:
                    if student["score"] >= 50:
                        print(f"{student['name']} is eligible for graduation.")
                    else:
                        print(f"{student['name']} is NOT eligible for graduation.")

        case 4:
            if not students:
                print("No students available.")
            else:
                top_student = max(students, key=lambda x: x["score"])
                print(f"Top Performer: {top_student['name']} with score {top_student['score']}")

        case 5:
            print("Exiting program. Goodbye!")
            break

        case _:
            print("Invalid choice. Please select between 1 and 5.")
```

```python
#Task 4.2:Eligibility checker
def check_graduation_eligibility(cgpa, outstanding_courses, is_active):
    if cgpa >= 2.5 and outstanding_courses == 0 and is_active:
        return True, "Eligible for graduation: CGPA requirement met, no outstanding courses, and student is active."

    else:
        reasons = []

        if cgpa < 2.5:
            reasons.append("CGPA is below 2.5")

        if outstanding_courses > 0:
            reasons.append("Outstanding courses remaining")

        if not is_active:
            reasons.append("Student is not active")

        return False, "Not eligible for graduation because: " + ", ".join(reasons)

cgpa_status = float(input("Enter your CGPA: "))
outstanding = int(input("Enter the outstanding courses: "))
active = ("Enter True or False: ")
status, message = check_graduation_eligibility(cgpa_status, outstanding, active)
print(status)   # True
print(message)

```

#Part 5":Advanced challenges(Optinal)

```python
#Task 5.1:Nested Data processing
students_scores = {
    "Amina": {"Math": 75, "English": 82, "Biology": 78},
    "Tunde": {"Math": 68, "English": 74, "Biology": 80},
    "Sadiq": {"Math": 85, "English": 90, "Biology": 88},
    "Zainab": {"Math": 72, "English": 69, "Biology": 75}
}

# Calculate average score for each student
averages = {}

for student, courses in students_scores.items():
    total = sum(courses.values())
    average = total / len(courses)
    averages[student] = average

# Display averages
print("Average scores:")
for student, avg in averages.items():
    print(f"{student}: {avg:.2f}")

# Find the first student who scored above 70 in all courses
first_student_above_70 = None

for student, courses in students_scores.items():
    if all(score > 70 for score in courses.values()):
        first_student_above_70 = student
        break

# Display result
if first_student_above_70:
    print(f"\nFirst student who scored above 70 in all courses: {first_student_above_70}")
else:
    print("\nNo student scored above 70 in all courses.")
```
