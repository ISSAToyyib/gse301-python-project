##2. Error Handling

```python
#2a and 2b
import csv

filename = "data.csv"
data = []

try:
    with open(filename, "r") as file:
        reader = csv.reader(file)

        for row in reader:
            if len(row) == 0:
                continue   # Skip empty lines
            try:
                value = float(row[0])
                data.append(value)
            except ValueError:
                print("Invalid value found and skipped:", row[0])

    if len(data) == 0:
        print("The file is empty or contains no valid numeric data.")
    else:
        print("Data read successfully:", data)

except FileNotFoundError:
    print("Error: The file does not exist.")
```

##3. Functions

```python
#3a and 3b
# Function to calculate total
def calculate_total(data):
    return sum(data)

# Function to calculate average
def calculate_average(data):
    if len(data) == 0:
        return 0
    return sum(data) / len(data)

# Function to calculate minimum
def calculate_minimum(data):
    if len(data) == 0:
        return None
    return min(data)

# Function to calculate maximum
def calculate_maximum(data):
    if len(data) == 0:
        return None
    return max(data)


# Example dataset
dataset = [65, 72, 80, 90, 55]

# Function calls
print("Total:", calculate_total(dataset))
print("Average:", calculate_average(dataset))
print("Minimum:", calculate_minimum(dataset))
print("Maximum:", calculate_maximum(dataset))
```

##4. Operators and Loops

```python
#4a. and 4b
# Dataset
data = [65, 72, 80, 90, 55]

# Initialize variables
total = 0
count = 0
minimum = data[0]
maximum = data[0]

# Loop through the dataset
for value in data:
    total = total + value       # Add each value to compute total
    count = count + 1           # Count data points

    # Compare to find minimum
    if value < minimum:
        minimum = value

    # Compare to find maximum
    if value > maximum:
        maximum = value

# Calculate average using arithmetic operator
average = total / count

# Display results
print("Total:", total)
print("Average:", average)
print("Minimum:", minimum)
print("Maximum:", maximum)
```

##7. OOP

```python
#7a and 7b
class DataSet:
    def __init__(self, filename):
        self.filename = filename
        self.data = []
        self.total = 0
        self.average = 0
        self.minimum = None
        self.maximum = None

    # Method to load numerical data from file
    def load_data(self):
        try:
            with open(self.filename, "r") as file:
                for line in file:
                    try:
                        value = float(line.strip())
                        self.data.append(value)
                    except ValueError:
                        pass   # Ignore non-numeric values

            if len(self.data) == 0:
                print("File is empty or contains no valid numeric data.")

        except FileNotFoundError:
            print("Error: File does not exist.")

    # Method to calculate statistics
    def calculate_statistics(self):
        if len(self.data) == 0:
            return

        self.total = 0
        self.minimum = self.data[0]
        self.maximum = self.data[0]

        for value in self.data:
            self.total += value

            if value < self.minimum:
                self.minimum = value

            if value > self.maximum:
                self.maximum = value

        self.average = self.total / len(self.data)

    # Method to display results
    def display_results(self):
        if len(self.data) == 0:
            print("No data to display.")
            return

        print("Total:", self.total)
        print("Average:", self.average)
        print("Minimum:", self.minimum)
        print("Maximum:", self.maximum)


# (b) Create an object and run analysis
dataset = DataSet("data.txt")
dataset.load_data()
dataset.calculate_statistics()
dataset.display_results()
```

##8. Saving results(File handling)

```python
class DataSet:
    def __init__(self, data_file, category_file):
        self.data_file = data_file
        self.category_file = category_file
        self.data = []
        self.total = 0
        self.average = 0
        self.minimum = None
        self.maximum = None
        self.unique_categories = set()

    def load_data(self):
        try:
            with open(self.data_file, "r") as file:
                for line in file:
                    try:
                        value = float(line.strip())
                        self.data.append(value)
                    except ValueError:
                        pass
        except FileNotFoundError:
            print("Numeric data file not found.")

    def load_categories(self):
        try:
            with open(self.category_file, "r") as file:
                for line in file:
                    category = line.strip()
                    if category != "":
                        self.unique_categories.add(category)
        except FileNotFoundError:
            print("Category file not found.")

    def calculate_statistics(self):
        if len(self.data) == 0:
            return

        self.total = 0
        self.minimum = self.data[0]
        self.maximum = self.data[0]

        for value in self.data:
            self.total += value
            if value < self.minimum:
                self.minimum = value
            if value > self.maximum:
                self.maximum = value

        self.average = self.total / len(self.data)

    def save_report(self, report_file):
        with open(report_file, "w") as file:
            file.write("DATA ANALYSIS REPORT\n")
            file.write("--------------------\n")
            file.write(f"Total: {self.total}\n")
            file.write(f"Average: {self.average}\n")
            file.write(f"Minimum: {self.minimum}\n")
            file.write(f"Maximum: {self.maximum}\n")
            file.write("\n")
            file.write(f"Total Unique Categories: {len(self.unique_categories)}\n")


# Create object and run analysis
dataset = DataSet("data.txt", "categories.txt")
dataset.load_data()
dataset.load_categories()
dataset.calculate_statistics()
dataset.save_report("report.txt")

print("Report saved successfully.")
```
